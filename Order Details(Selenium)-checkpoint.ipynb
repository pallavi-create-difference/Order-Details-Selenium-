```python
# import all the required libraries
import selenium
import pandas as pd
import time
from selenium import webdriver as wb 
import tqdm
from webdriver_manager.chrome import ChromeDriverManager
# initialise chromedriver
driver=wb.Chrome(ChromeDriverManager().install())
# opens flipkart through chrome using selenium
driver.get('https://www.flipkart.com/')
```

```python
# Put password and username within the single codes
username=''
password=''
# makes the username input in the login page
user_name=driver.find_element_by_xpath('/html/body/div[2]/div/div/div/div/div[2]/div/form/div[1]/input')
user_name.send_keys(username)
# enters the password in the login page
passWord=driver.find_element_by_xpath('/html/body/div[2]/div/div/div/div/div[2]/div/form/div[2]/input')
passWord.send_keys(password)
# clicks the login button
Login=driver.find_element_by_xpath('/html/body/div[2]/div/div/div/div/div[2]/div/form/div[3]/button')
Login.click()
# opens the user account
```

```python
# opens the order page of the user
order_page=driver.get('https://www.flipkart.com/account/orders?link=home_orders')
```

```python
# As 1 page doesn't contain all the orders ,to get all the order we need to scroll down  
lenofPage=driver.execute_script("window.scrollTo(0,document.body.scrollHeight);var lenofPage=document.body.scrollHeight;return lenofPage;")
```

```python
# this scrolls down to the end of the page containing all the order.
match=False
while(match==False):
    lastCount=lenofPage
    # Time gap to load each page
    time.sleep(3)
    lenofPage=driver.execute_script("window.scrollTo(0,document.body.scrollHeight);var lenofPage=document.body.scrollHeight;return lenofPage;")
    # ends the scroll action
    if lastCount==lenofPage:
        match=True
```

```python
# Scrapping of data starts from here
# A common class is found for all the orders
order=driver.find_elements_by_class_name('ShLswe')
```

```python
# empty list initialised
my_list=[]
for i in order:
    time.sleep(2)
    k=i.find_elements_by_class_name('_2WFi0x')
    time.sleep(0.5)
    for f in k:
        time.sleep(0.5)
# my_list contains the link of each order        
        my_list.append(f.get_property('href'))
```

```python
# a empty dictionary is created to store the product details.
dic1={'Name':None,'Price':None,'Order':None,'Delivary date':None}
List1=[]
```

```python
# looped through each order link to get the product details
for i in my_list:
    # get into individual order page
    driver.get(i)
    time.sleep(5)
    try:
        # extracts product details onlf if the product is delivered or is about to get deliver.
        delivery_date=driver.find_elements_by_class_name('_1i5nEe')[3].find_element_by_class_name('_3wS5ZT').text
    except:
        # if the delivery date not found, starts the loop again with another product link
        continue
    else:
        # extracts price and order date
        price=driver.find_element_by_class_name('_13fZDJ').text.split('\n')[0]
        order_date=driver.find_element_by_class_name('_3wS5ZT').text
        try:
            # scraping link of product page
            Product_link=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[2]/div/div[4]/div/div[1]/div[1]/div/div[2]/a')
        except:
            Product_link=None
        else:
            # opens the product page
            driver.get(Product_link.get_property('href'))
            time.sleep(2.4)
            Product_name=driver.find_element_by_class_name('_9E25nV').find_element_by_class_name('_35KyD6').text
            try:
                # extracts category of each product
                category=driver.find_elements_by_class_name('_1KHd47')[1].text
            except:
                category=None
        dic1={'Name':Product_name,'Price':price,'Order':order_date,'Delivary date':delivery_date,'Category':category}
        List1.append(dic1)
```

```python
# converts order details obtained into a dataframe.
data=pd.DataFrame(List1)
```
